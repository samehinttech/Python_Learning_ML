## Histograms in Python using Matplotlib
A histogram is a graphical representation of data that groups continuous number ranges into vertical bars, where the horizontal axis displays the number range and the vertical axis represents the frequency of data within each range. It allows for the visualization of the underlying frequency distribution of a dataset, helping to identify patterns such as normal distribution, outliers, and skewness. Unlike bar charts, histograms are specifically used for continuous or quantitative data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
# loc is the mean of simply that most of the grades are at the center
# scale is the standard deviation which shows how much the grades vary from the mean
# size is the number of grades to generate
grades = np.random.normal(loc=80, scale=10, size=100)

# Ensure grades are within 0-100, it clips any values below 0 to 0 and above 100 to 100 ( like min and max)
grades = np.clip(grades, 0, 100)

# Create histogram
plt.hist(grades,
         bins= 10,
         edgecolor='black',
         alpha=0.7,
         color='blue')
plt.title('Histogram of Student Grades')
plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.show()
```
